```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("../data/salesAnalysis.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df.columns = df.columns.str.strip()
```

```python
for col in df.columns:
    print(f"{col} has {df[col].nunique()} unique values")
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['Delivery_Time_Days'].value_counts()
```

```python
within_7_days = (df["Delivery_Time_Days"] <= 7).sum()

above_7_days = (df["Delivery_Time_Days"] > 7).sum()

print("Delivered ≤ 7 days:", within_7_days)
print("Delivered > 7 days:", above_7_days)
```

```python
df["Total_Sales"] = df["Total_Amount"]

df["Avg_Item_Price"] = df["Total_Amount"] / df["Quantity"]

df["Discount_Percentage"] = np.where(
    df["Unit_Price"] * df["Quantity"] == 0,
    0,
    df["Discount_Amount"] / (df["Unit_Price"] * df["Quantity"])
)
```

```python
df["Cost_Price"] = df["Unit_Price"] * 0.7
df["Profit"] = df["Total_Amount"] - df["Cost_Price"]
df["Profit_Margin"] = df["Profit"] / df["Total_Amount"]
```

```python
df["Cart_Size"] = df["Quantity"]
```

```python
df["Engagement_Score"] = df["Session_Duration_Minutes"] * df["Pages_Viewed"]

df["Pages_Per_Minute"] = np.where(
    df["Session_Duration_Minutes"] == 0,
    0,
    df["Pages_Viewed"] / df["Session_Duration_Minutes"]
)
```

```python
EXPECTED_DELIVERY_DAYS = 7

df["Is_Delayed"] = (df["Delivery_Time_Days"] > EXPECTED_DELIVERY_DAYS).astype(int)
```

```python
df["Rating_Category"] = pd.cut(
    df["Customer_Rating"],
    bins=[0, 2, 4, 5],
    labels=["Low", "Medium", "High"]
)
```

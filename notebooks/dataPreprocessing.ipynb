```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Total_Amount"] = (df["Unit_Price"] * df["Quantity"]) - df["Discount_Amount"]
df["Total_Sales"] = df["Total_Amount"]
```

```python
df["Avg_Item_Price"] = np.where(
    df["Quantity"] != 0,
    df["Total_Amount"] / df["Quantity"],
    0
)
```

```python
gross_amount = df["Unit_Price"] * df["Quantity"]

df["Discount_Percentage"] = np.where(
    gross_amount != 0,
    (df["Discount_Amount"] / gross_amount) * 100,
    0
)
```

```python
df["Cost_Price"] = df["Unit_Price"] * df["Quantity"] * 0.70
```

```python
df["Profit"] = df["Total_Amount"] - df["Cost_Price"]
```

```python
df["Profit_Margin"] = np.where(
    df["Total_Amount"] != 0,
    (df["Profit"] / df["Total_Amount"]) * 100,
    0
)
```

```python
df["Engagement_Score"] = (
    df["Session_Duration_Minutes"] *
    df["Pages_Viewed"]
)
```

```python
df["Pages_Per_Minute"] = np.where(
    df["Session_Duration_Minutes"] != 0,
    df["Pages_Viewed"] / df["Session_Duration_Minutes"],
    0
)
```

```python
df["Is_Delayed"] = df["Delivery_Time_Days"] > 7
```

```python
df["Rating_Category"] = pd.cut(
    df["Customer_Rating"],
    bins=[0, 2, 3.5, 5],
    labels=["Poor", "Average", "Excellent"],
    include_lowest=True
)
```

# Data Preprocessing and Feature Engineering

Before beginning the Exploratory Data Analysis (EDA), the dataset was preprocessed to ensure data quality and to create meaningful business metrics that would provide deeper insights during analysis.

## Date Data Type Conversion

The `Date` column was initially stored as an **object (string)** data type. It was converted to the **datetime** format to enable time-based analysis. This conversion allows for extracting useful information such as year, month, day, and weekday, making it easier to identify seasonal trends, monthly sales patterns, customer purchasing behavior over time, and perform chronological sorting and filtering.

## Feature Engineering

Several new features were created from the existing data to enhance the analytical value of the dataset. These engineered features represent key business metrics commonly used in e-commerce and retail analytics.

- **Total_Amount** was generated to represent the final amount paid by the customer after applying discounts. This serves as the primary revenue metric for each transaction.

- **Total_Sales** was created as a business-friendly representation of the transaction value, making reporting and dashboard development more intuitive.

- **Avg_Item_Price** was calculated to understand the average selling price of products within each order. This helps analyze purchasing behavior and pricing strategies.

- **Discount_Percentage** was generated to measure the proportion of discount offered on each purchase, allowing analysis of discount effectiveness and promotional impact.

- **Cost_Price** was estimated because actual procurement costs were not available in the dataset. Assuming a fixed percentage of the selling price enables profitability analysis in a realistic manner.

- **Profit** was created to estimate the earnings generated from each transaction after considering the estimated product cost.

- **Profit_Margin** was introduced to evaluate how efficiently each sale generates profit relative to its revenue, making it easier to compare profitability across different orders.

- **Engagement_Score** combines customer browsing duration and page views into a single metric that reflects the overall level of customer interaction with the website.

- **Pages_Per_Minute** measures browsing efficiency by indicating how actively customers navigate the website during their shopping sessions.

- **Is_Delayed** was created as a binary indicator to identify whether an order exceeded the expected delivery time. This simplifies the analysis of delivery performance and logistics efficiency.

- **Rating_Category** groups customer ratings into broader categories such as Poor, Average, and Excellent. Categorizing ratings makes it easier to analyze customer satisfaction trends and compare them across different customer segments.

## Summary

These preprocessing and feature engineering steps transformed the raw transactional data into a more informative and analysis-ready dataset. The newly created features provide valuable insights into sales performance, profitability, customer behavior, engagement, delivery efficiency, and customer satisfaction, enabling a more comprehensive and business-oriented exploratory data analysis.
